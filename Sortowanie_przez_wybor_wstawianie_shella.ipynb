# Sortowanie przez wybór (Selection Sort)
Sortowanie przez wybór to prosty algorytm sortowania polegający na wielokrotnym wybieraniu najmniejszego elementu z nieposortowanej części tablicy i zamianie go z pierwszym elementem tej części.

Algorytm działa według następującej zasady:
1. Znajdź najmniejszy element w całej tablicy.
2. Zamień go z pierwszym elementem tablicy.
3. Następnie znajdź najmniejszy element w pozostałej części tablicy.
4. Zamień go z drugim elementem.
5. Powtarzaj proces aż do końca tablicy.

Po każdym przebiegu jeden element znajduje się już na właściwej pozycji.

## Algorytm sortowania przez wybór w C++

```python
#include <iostream>
using namespace std;

void selectionSort(int tab[], int n) {
    for(int i = 0; i < n - 1; i++) {
        int minIndex = i;

        for(int j = i + 1; j < n; j++) {
            if(tab[j] < tab[minIndex]) {
                minIndex = j;
            }
        }

        swap(tab[i], tab[minIndex]);
    }
}

int main() {
    int tab[] = {64, 25, 12, 22, 11};
    int n = 5;

    selectionSort(tab, n);

    for(int i = 0; i < n; i++)
        cout << tab[i] << " ";

    return 0;
}
```

# Sortowanie przez wstawianie (Insertion Sort)

Sortowanie przez wstawianie (ang. Insertion Sort) to prosty algorytm sortowania polegający na stopniowym budowaniu posortowanej części tablicy poprzez wstawianie kolejnych elementów w odpowiednie miejsce.

Algorytm działa podobnie jak układanie kart w ręce podczas gry. Każdą nową kartę wstawiamy w odpowiednie miejsce wśród już uporządkowanych kart.

Zasada działania
1. Przyjmujemy, że pierwszy element tablicy jest już posortowany.
2. Pobieramy kolejny element.
3. Porównujemy go z elementami znajdującymi się wcześniej w tablicy.
4. Przesuwamy większe elementy o jedną pozycję w prawo.
5. Wstawiamy element w odpowiednie miejsce.
6. Powtarzamy proces dla kolejnych elementów tablicy.

Algorytm jest szczególnie wydajny dla małych lub częściowo posortowanych zbiorów danych.

## Algorytm sortowania przez wstawianie w C++:

```python
#include <iostream>
using namespace std;

void insertionSort(int tab[], int n) {

    for(int i = 1; i < n; i++) {

        int key = tab[i];
        int j = i - 1;

        while(j >= 0 && tab[j] > key) {
            tab[j + 1] = tab[j];
            j--;
        }

        tab[j + 1] = key;
    }
}

int main() {

    int tab[] = {8,3,5,2,9};
    int n = 5;

    insertionSort(tab,n);

    for(int i=0;i<n;i++)
        cout<<tab[i]<<" ";

    return 0;
}
```

# Sortowanie metodą Shella (Shell Sort)

Sortowanie metodą Shella jest ulepszeniem algorytmu sortowania przez wstawianie (Insertion Sort). Zostało zaproponowane w 1959 roku przez Donalda Shella.

Główną ideą algorytmu jest porównywanie elementów oddalonych od siebie o pewną wartość zwaną krokiem (gap). Dzięki temu elementy mogą szybciej przemieszczać się na właściwe miejsca w tablicy.

Algorytm wykonuje kilka przebiegów sortowania przez wstawianie, ale dla różnych odstępów między elementami.

Zasada działania
1. Wybieramy początkowy krok gap (najczęściej n/2).
2. Sortujemy elementy oddalone o gap pozycji metodą wstawiania.
3. Zmniejszamy wartość gap.
4. Powtarzamy proces aż gap = 1.
5. Ostatni przebieg jest klasycznym sortowaniem przez wstawianie.

Dzięki wcześniejszym operacjom tablica jest już częściowo uporządkowana, co przyspiesza ostatni etap.

## Algorytm sortowania metodą Shella w C++

```python
#include <iostream>
using namespace std;

void shellSort(int tab[], int n) {

    for(int gap = n/2; gap > 0; gap /= 2) {

        for(int i = gap; i < n; i++) {

            int temp = tab[i];
            int j;

            for(j = i; j >= gap && tab[j-gap] > temp; j -= gap)
                tab[j] = tab[j-gap];

            tab[j] = temp;
        }
    }
}

int main() {

    int tab[] = {23,12,1,8,34,54,2,3};
    int n = 8;

    shellSort(tab,n);

    for(int i=0;i<n;i++)
        cout<<tab[i]<<" ";

}
```

# Zadanie 4
Zmodyfikuj program Selection Sort tak, aby policzył liczbę wykonanych porównań.

```python
#include <iostream>
using namespace std;

void selectionSort(int tab[], int n) {
    int iterations = 0;
    for(int i = n - 1; i >= 0; i--) {
        int minIndex = i;

        for(int j = i - 1; j >= 0; j--) {
            if(tab[j] < tab[minIndex]) {
                minIndex = j;
                iterations++;
                cout << "Iteracja: " << iterations << ", minIndex: " << minIndex << ", tab[minIndex]: " << tab[minIndex] << endl;
            }
        }

        swap(tab[i], tab[minIndex]);
    }
    cout << "Liczba iteracji: " << iterations << endl;
}

int main() {
    
    cout << "Podaj liczbe elementow: " << endl;
    int n;
    cin >> n;
    int tab[n];
    cout << "Podaj elementy: "<< endl;
    for(int i = 0; i < n; i++){
        cin >> tab[i];
    }
    selectionSort(tab, n);

    for(int i = 0; i < n; i++)
        cout << tab[i] << " ";

    return 0;
}
```

# Zadanie 5
Zmierz czas sortowania przez wybór tablicy 10 000 elementów przy użyciu biblioteki <chrono>.

```python
#include <iostream>
#include <chrono>
#include <thread>
using namespace std;

void selectionSort(int tab[], int n) {
    for(int i = n - 1; i >= 0; i--) {
        int minIndex = i;

        for(int j = i - 1; j >= 0; j--) {
            if(tab[j] < tab[minIndex]) {
                minIndex = j;
                
            }
        }

        swap(tab[i], tab[minIndex]);
    }
}

int main() {
    auto start = chrono::steady_clock::now();
    this_thread::sleep_for(std::chrono::milliseconds(100));
    int const n = 10000;
    int tab[n];
    selectionSort(tab, n);       

    auto end = std::chrono::steady_clock::now();
    auto diff = end - start;
    cout << "Czas: " << std::chrono::duration_cast<std::chrono::milliseconds>(diff).count() 
              << " ms" << std::endl;
    return 0;
}
```
